# Seperate test and train data

```julia
df = DataFrame(Date=reverse(df[!,"Date"]),Open=reverse(df[!,"Open"]),High=reverse(df[!,"High"]),Low=reverse(df[!,"Low"]),Close=reverse(df[!,"Close"]),Volume=reverse(df[!,"Volume"]))
```

```julia
describe(df)[!,"max"][2]
```

```julia
describe(df)
```

```julia
for i in 1:(size(df[!,"Open"],1))
    df[!,"Open"][i] = (df[!,"Open"][i]-describe(df)[!,"min"][2])/(describe(df)[!,"max"][2]-describe(df)[!,"min"][2])
end
describe(df[!,"Open"])
```

```julia
for i in 1:(size(df[!,"Close"],1))
    df[!,"Close"][i] = (df[!,"Close"][i]-describe(df)[!,"min"][5])/(describe(df)[!,"max"][5]-describe(df)[!,"min"][5])
end
describe(df[!,"Close"])
```

```julia
for i in 1:(size(df[!,"High"],1))
    df[!,"High"][i] = (df[!,"High"][i]-describe(df)[!,"min"][3])/(describe(df)[!,"max"][3]-describe(df)[!,"min"][3])
end
describe(df[!,"High"])
```

```julia
for i in 1:(size(df[!,"Low"],1))
    df[!,"Low"][i] = (df[!,"Low"][i]-describe(df)[!,"min"][4])/(describe(df)[!,"max"][4]-describe(df)[!,"min"][4])
end
describe(df[!,"Low"])
```

```julia
df[!,"Volume"] = df[!,"Volume"]/(10^6)
for i in 1:(size(df[!,"Volume"],1))
    df[!,"Volume"][i] = (df[!,"Volume"][i]-describe(df)[!,"min"][6])/(describe(df)[!,"max"][6]-describe(df)[!,"min"][6])
end
describe(df[!,"Volume"])
```
